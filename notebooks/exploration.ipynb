### data_exploration.ipynb: 
goal: explore ufc fights dataset to understand shape, columns, and make visualizations

```python
import pandas as pd
import numpy as np
```

```python
# some settings:
pd.set_option('display.max_columns', 100)
```

```python
ufc = pd.read_csv('../data/ufc-master.csv')
ufc.head()
```

```python
ufc.shape
```

```python
ufc.dtypes.value_counts()
```

```python
print(ufc.columns.tolist())
```

```python
# show columns with > 5% missing
missing = ufc.isnull().mean().sort_values(ascending = False)
missing[missing > 0.05]
```

```python
ufc.dtypes.value_counts()
```

```python
# list object columns
obj_cols = ufc.select_dtypes(include = 'object').columns.tolist()
print(obj_cols)

# preview value counts
for col in obj_cols:
    print(f"\n--- {col} ---")
    print(ufc[col].value_counts(dropna = False).head(10))
```
