### data_exploration.ipynb: 
exploring ufc fights dataset with a goal to understand shape, columns, and make visualizations

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# some settings:
pd.set_option('display.max_columns', 100)
sns.set(style = 'whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
ufc = pd.read_csv('../data/ufc-master.csv')
ufc.head()
```

```python
ufc.shape
```

```python
ufc.dtypes.value_counts()
```

```python
print(ufc.columns.tolist())
```

```python
# show columns with > 5% missing
missing = ufc.isnull().mean().sort_values(ascending = False)
missing[missing > 0.05]
```

```python
ufc.dtypes.value_counts()
```

```python
# list object columns
obj_cols = ufc.select_dtypes(include = 'object').columns.tolist()
print(obj_cols)

# preview value counts
for col in obj_cols:
    print(f"\n--- {col} ---")
    print(ufc[col].value_counts(dropna = False).head(10))
```

### Data Visualizations:

```python
sns.countplot(data = ufc, x = 'Winner', hue = 'Winner', palette = {'Red': 'red', 'Blue': 'blue'})
plt.title('winner distribution: red vs blue')
plt.show()
```

```python
sns.countplot(data = ufc, y = 'Finish', order = ufc['Finish'].value_counts().index)
plt.title('fight finish types')
plt.xlabel('count')
plt.ylabel('finish')
plt.show()
```

```python
sns.countplot(data = ufc, x = 'TitleBout')
plt.title('how many fights were title bouts')
plt.show()
```

```python
sns.countplot(data = ufc, y = 'WeightClass', order = ufc['WeightClass'].value_counts().index)
plt.title('weight class distribution')
plt.xlabel('count')
plt.ylabel('weight class')
plt.show()
```

```python
sns.histplot(ufc['AgeDif'].dropna(), bins = 30, kde = True)
plt.axvline(0, color = 'black', linestyle = '--')
plt.title('age difference (red - blue)')
plt.xlabel('age difference')
plt.show()
```

```python
ax = sns.boxplot(data = ufc, x = 'Winner', y = 'ReachDif', order = ['Red', 'Blue'], 
                palette = {'Red': '#d62728', 'Blue': '#1f77b4'}, showfliers = False)  # hides extreme outliers for cleaner look

ax.set_title('reach difference vs winner')
ax.set_xlabel('winner')
ax.set_ylabel('reach difference (red - blue)')
ax.axhline(0, color = 'black', linestyle = '--', linewidth = 1)

plt.show()
```
